```python
# Crop Production Analysis

## Introduction
Agriculture plays a critical role in food security and economic development.  
This project analyzes crop production data across different states, seasons, and years to identify patterns, trends, and key insights.

## Objectives
- Identify the most produced crops
- Analyze production trends over time
- Compare production across seasons and regions
- Evaluate crop efficiency using yield
```

```python
# ================================
# 1. Import Libraries
# ================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")


# ================================
# 2. Load Dataset
# ================================
# Since your dataset is in the SAME folder:
df = pd.read_csv("crop_data.csv")

# Preview data
print("First 5 rows:")
print(df.head())


# ================================
# 3. Explore Data
# ================================
print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())


# ================================
# 4. Data Cleaning
# ================================
# Drop missing values
df = df.dropna()

# Ensure correct data type
df["Crop_Year"] = df["Crop_Year"].astype(int)


# ================================
# 5. Feature Engineering (Yield)
# ================================
df["Yield"] = df["Production"] / df["Area"]

print("\nData after adding Yield:")
print(df.head())


# ================================
# 6. Top Crops by Production
# ================================
top_crops = df.groupby("Crop")["Production"].sum().sort_values(ascending=False).head(10)

plt.figure()
top_crops.plot(kind="bar")
plt.title("Top 10 Crops by Production")
plt.xlabel("Crop")
plt.ylabel("Production")
plt.xticks(rotation=45)
plt.show()


# ================================
# 7. Production Over Time
# ================================
yearly_production = df.groupby("Crop_Year")["Production"].sum()

plt.figure()
yearly_production.plot()
plt.title("Total Production Over Years")
plt.xlabel("Year")
plt.ylabel("Production")
plt.show()


# ================================
# 8. Production by Season
# ================================
season_prod = df.groupby("Season")["Production"].sum()

plt.figure()
season_prod.plot(kind="bar")
plt.title("Production by Season")
plt.xlabel("Season")
plt.ylabel("Production")
plt.xticks(rotation=45)
plt.show()


# ================================
# 9. Top States by Production
# ================================
top_states = df.groupby("State_Name")["Production"].sum().sort_values(ascending=False).head(10)

plt.figure()
top_states.plot(kind="bar")
plt.title("Top 10 States by Production")
plt.xlabel("State")
plt.ylabel("Production")
plt.xticks(rotation=45)
plt.show()


# ================================
# 10. Yield Analysis (Efficiency)
# ================================
yield_by_crop = df.groupby("Crop")["Yield"].mean().sort_values(ascending=False).head(10)

plt.figure()
yield_by_crop.plot(kind="bar")
plt.title("Top Crops by Yield (Efficiency)")
plt.xlabel("Crop")
plt.ylabel("Yield")
plt.xticks(rotation=45)
plt.show()


# ================================
# 11. Correlation Heatmap
# ================================
plt.figure()
sns.heatmap(df[["Area", "Production", "Yield"]].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()


# ================================
# 12. Bonus: Single Crop Analysis (Rice Example)
# ================================
rice_data = df[df["Crop"] == "Rice"]

rice_yearly = rice_data.groupby("Crop_Year")["Production"].sum()

plt.figure()
rice_yearly.plot()
plt.title("Rice Production Over Time")
plt.xlabel("Year")
plt.ylabel("Production")
plt.show()
```
